```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members! 

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()

    question10 = """
        SELECT facility_name, SUM(revenue) AS total_revenue
        FROM (SELECT f.name as facility_name,             
                     CASE WHEN b.memid!=0 THEN b.slots*f.membercost
                          ELSE b.slots*f.guestcost
                          END AS revenue
              FROM Bookings b
              LEFT JOIN Facilities f
              ON b.facid=f.facid)     
        GROUP BY 1
        HAVING SUM(revenue)<1000
        ORDER BY 2 DESC;
        """
    cur.execute(question10)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn:
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q11: Produce a report of members and who recommended them 
# in alphabetic surname,firstname order

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()

    question11 = """
    SELECT t1.surname,t1.firstname, t2.surname, t2.firstname
FROM Members t1, Members t2
WHERE t1.recommendedby=t2.memid AND t2.memid!=0
ORDER BY 1,2;

        """
    cur.execute(question11)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
       
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q12 Find the facilities with their usage by member, but not guests
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()

    question12 = """
        SELECT memid,
       SUM(CASE WHEN facid=0 THEN 1
            ELSE 0 END) AS "Tennis Court 1",
       SUM(CASE WHEN facid=1 THEN 1
            ELSE 0 END) AS "Tennis Court 2",
       SUM(CASE WHEN facid=2 THEN 1
            ELSE 0 END) AS "Badminton Court",
       SUM(CASE WHEN facid=3 THEN 1
            ELSE 0 END) AS "Table Tennis",
       SUM(CASE WHEN facid=4 THEN 1
            ELSE 0 END) AS "Massage Room 1",
       SUM(CASE WHEN facid=5 THEN 1
            ELSE 0 END) AS "Massage Room 2",
       SUM(CASE WHEN facid=6 THEN 1
            ELSE 0 END) AS "Squash Court",
       SUM(CASE WHEN facid=7 THEN 1
            ELSE 0 END) AS "Snooker Table",
       SUM(CASE WHEN facid=8 THEN 1
            ELSE 0 END) AS "Pool Table"
FROM Bookings
WHERE memid!=0
GROUP BY memid
        """
    cur.execute(question12)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
       
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q13 Find the facilities with their usage by month, but not guests
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()

    question13 = """
        SELECT strftime('%m', starttime) AS Month,
       SUM(CASE WHEN facid=0 THEN 1
            ELSE 0 END) AS "Tennis Court 1",
       SUM(CASE WHEN facid=1 THEN 1
            ELSE 0 END) AS "Tennis Court 2",
       SUM(CASE WHEN facid=2 THEN 1
            ELSE 0 END) AS "Badminton Court",
       SUM(CASE WHEN facid=3 THEN 1
            ELSE 0 END) AS "Table Tennis",
       SUM(CASE WHEN facid=4 THEN 1
            ELSE 0 END) AS "Massage Room 1",
       SUM(CASE WHEN facid=5 THEN 1
            ELSE 0 END) AS "Massage Room 2",
       SUM(CASE WHEN facid=6 THEN 1
            ELSE 0 END) AS "Squash Court",
       SUM(CASE WHEN facid=7 THEN 1
            ELSE 0 END) AS "Snooker Table",
       SUM(CASE WHEN facid=8 THEN 1
            ELSE 0 END) AS "Pool Table"
FROM Bookings
WHERE memid!=0
GROUP BY 1
        """
    cur.execute(question13)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
       
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
